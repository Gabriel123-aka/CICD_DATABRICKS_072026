##Revocar rol de Data Engineers (DEs)

```python
%sql
-- CATALOGO
REVOKE ALL PRIVILEGES ON CATALOG retail_dev FROM `DEs`;

-- PERMISOS ESQUEMAS
REVOKE CREATE TABLE, USE SCHEMA ON SCHEMA retail_dev.bronze FROM `DEs`;
REVOKE CREATE TABLE, USE SCHEMA ON SCHEMA retail_dev.silver FROM `DEs`;
REVOKE CREATE TABLE, USE SCHEMA ON SCHEMA retail_dev.golden FROM `DEs`;

-- PERMISOS TABLAS BRONZE
REVOKE SELECT ON TABLE retail_dev.bronze.clientes FROM `DEs`;
REVOKE SELECT ON TABLE retail_dev.bronze.productos FROM `DEs`;
REVOKE SELECT ON TABLE retail_dev.bronze.ordenes FROM `DEs`;
REVOKE SELECT ON TABLE retail_dev.bronze.detalle_ordenes FROM `DEs`;
REVOKE SELECT ON TABLE retail_dev.bronze.inventario FROM `DEs`;

-- PERMISOS TABLAS SILVER
REVOKE SELECT ON TABLE retail_dev.silver.ventas_detalle_limpias FROM `DEs`;
REVOKE SELECT ON TABLE retail_dev.silver.inventario_limpio FROM `DEs`;

-- PERMISOS TABLA GOLDEN
REVOKE SELECT ON TABLE retail_dev.golden.resumen_retail FROM `DEs`;
```
