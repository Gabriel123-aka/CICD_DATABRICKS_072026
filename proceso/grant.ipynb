##Rol Data Engineers (DEs)

```python
%sql
-- CATÁLOGO
GRANT ALL PRIVILEGES ON CATALOG retail_dev TO `DEs`;

-- ESQUEMAS
GRANT CREATE TABLE, USE SCHEMA ON SCHEMA retail_dev.bronze TO `DEs`;
GRANT CREATE TABLE, USE SCHEMA ON SCHEMA retail_dev.silver TO `DEs`;
GRANT CREATE TABLE, USE SCHEMA ON SCHEMA retail_dev.golden TO `DEs`;

-- TABLAS BRONZE
GRANT SELECT ON TABLE retail_dev.bronze.clientes TO `DEs`;
GRANT SELECT ON TABLE retail_dev.bronze.productos TO `DEs`;
GRANT SELECT ON TABLE retail_dev.bronze.ordenes TO `DEs`;
GRANT SELECT ON TABLE retail_dev.bronze.detalle_ordenes TO `DEs`;
GRANT SELECT ON TABLE retail_dev.bronze.inventario TO `DEs`;

-- TABLAS SILVER
GRANT SELECT ON TABLE retail_dev.silver.ventas_detalle_limpias TO `DEs`;
GRANT SELECT ON TABLE retail_dev.silver.inventario_limpio TO `DEs`;

-- TABLA GOLDEN
GRANT SELECT ON TABLE retail_dev.golden.resumen_retail TO `DEs`;
```

```python
%sql
-- EXTERNAL LOCATIONS
GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-metastore`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-raw`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-bronze`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-silver`
TO `DEs`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-golden`
TO `DEs`;
```
